```python
%sql
CREATE OR REPLACE TABLE invt.Factorderstatus
SELECT 
dos.orderstatusID,
dloc.locationID,
dd.DateID,
bmd.ProductQuantity,
dp.ProductID
FROM invt.order_status as os , invt.bharat_mart_dataset as bmd 
INNER JOIN invt.dim_orderstatus dos ON os.orderstatus = dos.orderstatus
INNER JOIN invt.dim_location dloc On os.customeraddress = dloc.location
LEFT JOIN invt.dim_date dd on CAST(bmd.OrderDate AS DATE ) = dd.Date
INNER JOIN invt.dim_product dp ON bmd.ProductName = dp.ProductName;

-- SELECT * FROM INVT.Factorderstatus
```

```python
%sql
CREATE OR REPLACE TABLE  invt.FactProductPurchase as
SELECT 
ph.Price,
ph.Quantity,
dp.ProductID,
ws.WholesalerID,
dr.RetailerID,
cast(ph.PurchaseDate as date) as purchaseDate
from invt.purchasehistory as ph
inner join invt.dim_product as dp ON ph.ProductName = dp.ProductName
inner join invt.dim_wholesales as ws on ws.WholeSalerName= ph.WholeSalerName
inner join invt.dim_retailer as dr ON ph.RetailerName = dr.RetailerName
left join invt.dim_date as dd on cast(ph.PurchaseDate as date) = dd.Date;

select * from invt.factproductpurchase
```

```python
%sql
CREATE OR REPLACE TABLE invt.FactInventoryProduct
SELECT 
    dc.CustomerID,
    dp.ProductID,
    dr.RetailerID,
    da.AgentID,
    dw.WareHouseID,
    dpy.PaymentID,
    dd.DateID AS mfdDateID,
    dat.DateID AS expDateID,
    ddat.DateID AS orderDateID,
    dte.DateID AS deliveryDateID,
    rmb.CostPrice,
    rmb.SellingPrice,
    rmb.Discount,
    rmb.Price,
    rmb.ProductQuantity
FROM invt.bharat_mart_dataset rmb
INNER JOIN invt.dim_customer dc ON rmb.CustomerName = dc.CustomerName
INNER JOIN invt.dim_product dp ON rmb.ProductName = dp.ProductName
INNER join invt.dim_retailer dr ON rmb.RetailerName = dr.RetailerName
INNER join invt.dim_agents da On rmb.AgentName = da.AgentName
INNER JOIn invt.dim_warehouse dw ON rmb.WareHouseName = dw.warehousename
INNER JOIN invt.dim_paymentmode dpy ON rmb.PaymentMode = dpy.paymentmode
LEFT JOIN invt.dim_date dd ON CAST(rmb.MFDDate AS DATE) = dd.Date
LEFT JOIN invt.dim_date dat ON CAST(rmb.EXPDate AS DATE) = dat.Date
LEFT JOIN invt.dim_date dte ON CAST(rmb.DeliveryDate AS DATE) = dte.Date
LEFT JOIN invt.dim_date ddat ON CAST(rmb.OrderDate AS DATE) = ddat.Date;

select * from invt.FactInventoryProduct
```

```python
%sql
CREATE OR REPLACE TABLE invt.Factproductstatus
select 
ss.ProductQuantity,
ss.Price,
ps.productstatusID,
dp.ProductID,
dw.WareHouseID
from invt.samplestock_csv as ss
join invt.dim_productstatus as ps on ss.ProductStatus = ps.ProductStatus
join invt.dim_product as dp on ss.ProductName = dp.ProductName
join invt.dim_warehouse as dw on ss.WareHouseName = dw.warehousename;

select * from invt.Factproductstatus
```
