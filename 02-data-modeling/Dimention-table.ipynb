```python
%sql
-- """ Creating Customer Dimention Table
-- """
CREATE OR REPLACE TABLE invt.DIM_Customer
AS
WITH CTE_Customer AS (
    SELECT DISTINCT
        CustomerName,
        CustomerMail,
        CustomerGender,
        CustomerContact,
        CustomerCity,
        CustomerAge,
        CustomerAddress
    FROM invt.bharat_mart_dataset
)
SELECT
    row_number() OVER (ORDER BY 1) AS CustomerID,
    CustomerName,
    CustomerMail,
    CustomerGender,
    CustomerContact,
    CustomerCity,
    CustomerAge,
    CustomerAddress
FROM CTE_Customer

UNION ALL

SELECT
    -1,
    'Unknown',
    'Unknown',
    'Unknown',
    -1,
    'Unknown',
    -1,
    'Unknown';

```

```python
%sql
-- """ 
-- Creating Product Dimention Table       
-- """

CREATE  OR REPLACE TABLE invt.DIM_Product
AS
WITH CTE_Product AS (
    SELECT DISTINCT
        ProductCategory,
        ProductName,
        productquantity
    FROM invt.bharat_mart_dataset
)
SELECT
    row_number() OVER (ORDER BY 1) AS ProductID,
    ProductCategory,
        ProductName ,
        productquantity
FROM cte_product

UNION ALL

SELECT
    -1 AS ProductID,
    'Unknown' AS ProductCategory,
        'Unknown' AS ProductName ,
    -1 as productquantity ;

    select * from invt.DIM_Product
```

```python
# """
# creating dim date table
# """
import pandas as pd
from pyspark.sql.functions import year, month, dayofmonth, date_format, to_timestamp

start_date = "2015-06-01"
end_date = "2024-10-30"

date_range = pd.date_range(start=start_date, end=end_date)
date_df = pd.DataFrame({'Date': date_range})

df1 = spark.createDataFrame(date_df)
df1 = df1.withColumn("Date", to_timestamp(df1["Date"]))
df1 = df1.withColumn("Year", year(df1["Date"]))
df1 = df1.withColumn("Month", month(df1["Date"]))
df1 = df1.withColumn("DayofMonth", dayofmonth(df1["Date"]))
df1 = df1.withColumn("DateID", date_format(df1["Date"], "yyyyMMdd"))

df1.createOrReplaceTempView("DateDetails")
df1.write.mode("overwrite").saveAsTable("invt.dim_date")

```
