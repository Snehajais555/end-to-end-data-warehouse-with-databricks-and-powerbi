```python
%sql
-- """ 
--   Fact Table data validation """
SELECT
 COUNT( DISTINCT CustomerID) as CUSTOMER_couunt,
 COUNT( DISTINCT ProductID) as PRODUCT_count,
 COUNT(DISTINCT RetailerID) as RETAILER_count,
 COUNT(DISTINCT AgentID) as AGENT_count,
 COUNT(DISTINCT WareHouseID) as WAREHOUSE_count,
 COUNT(DISTINCT PaymentID) as PAYMENT_count,
 COUNT(DISTINCT CostPrice) AS costprice_count,
 COUNT( DISTINCT SellingPrice) AS sellingprice_count,
 COUNT(DISTINCT Discount) AS discount_count,
 COUNT(DISTINCT Price) AS price_count

FROM invt.facttable;
```
